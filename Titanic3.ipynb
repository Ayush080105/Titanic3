```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic3.csv')
df.sample(5)
```

```python
df=df.drop(columns=['name','cabin','body','boat','home.dest','ticket'],axis=1)
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['survived']),df['survived'],test_size=0.2,random_state=0)
```

```python
x_test = x_test.drop(x_test.index[253])
x_train=x_train.drop(x_train.index[253])
y_test=y_test.drop(y_test.index[253])
y_train=y_train.drop(y_train.index[253])
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer()
freq=SimpleImputer(strategy='most_frequent')
x_train_age=impute.fit_transform(x_train[['age']])
x_train_embarked=freq.fit_transform(x_train[['embarked']])
x_test_age=impute.fit_transform(x_test[['age']])
x_test_embarked=freq.fit_transform(x_test[['embarked']])

```

```python
from sklearn.preprocessing import OneHotEncoder
ohe_sex =OneHotEncoder(handle_unknown='ignore')
ohe_embarked =OneHotEncoder(handle_unknown='ignore')
x_train_sex =ohe_sex.fit_transform(x_train[['sex']])
x_train_embarked =ohe_embarked.fit_transform(x_train_embarked)
x_test_sex =ohe_sex.fit_transform(x_test[['sex']])
x_test_embarked =ohe_embarked.fit_transform(x_test_embarked)
```

```python
x_train_sex=x_train_sex.toarray()
x_train_embarked=x_train_embarked.toarray()
x_test_sex=x_test_sex.toarray()
x_test_embarked=x_test_embarked.toarray()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_age=sc.fit_transform(x_train_age)
x_test_age=sc.fit_transform(x_test_age)
x_train_fare=sc.fit_transform(x_train[['fare']])
x_test_fare=sc.fit_transform(x_test[['fare']])
```

```python
x_train_rem=x_train.drop(columns=['sex','age','embarked','fare'],axis=1)
x_test_rem=x_test.drop(columns=['sex','age','embarked','fare'],axis=1)
```

```python
x_test_rem.shape
```

```python
x_test_new=np.concatenate((x_test_rem,x_test_age,x_test_fare,x_test_sex,x_test_embarked),axis=1)
x_train_new=np.concatenate((x_train_rem,x_train_age,x_train_fare,x_train_sex,x_train_embarked),axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train_new,y_train)
```

```python
y_pred=dtc.predict(x_test_new)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```
